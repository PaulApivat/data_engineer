```python
import duckdb
import pandas as pd
```

```python
# Define the path to your DuckDB database file
db_path = '../data/bronze/raw_data.db'

# Connect to the DuckDB database
con = duckdb.connect(database=db_path)

# Execute SQL queries
result = con.execute("SELECT * FROM eth_emissions")

# Fetch and display data
data = result.fetchall()
print(data)

# Turn to dataframe
columns = [desc[0] for desc in result.description]
df = pd.DataFrame(data, columns=columns)
print(df)

# Close the connection
con.close()
```

```python
# Define the path to your DuckDB database file
db_path = '../data/silver/transform_data.db'

# Connect to the DuckDB database
con = duckdb.connect(database=db_path)

# Execute SQL queries
result = con.execute("SELECT * FROM eth_emissions_silver")

# Fetch and display data
silver_data = result.fetchall()
print(type(silver_data))
print(silver_data)

# Turn list of tuple into a dataframe
columns = ["datetime", "net_emission_eth", "total_net_emission_eth", "token_symbol", "protocol", "category", "layer", "measure"]
silver_df = pd.DataFrame(silver_data, columns=columns)
print(silver_df)

# Close the connection
con.close()
```

```python
# Define the path to your DuckDB database file
db_path = '../data/gold/consumption_data.db'

# Connect to the DuckDB database
con = duckdb.connect(database=db_path)

# Execute SQL queries
result = con.execute("SELECT * FROM eth_emissions_gold")

# Fetch and display data
gold_data = result.fetchall()
#print(gold_data)

# Turn to dataframe
columns = [desc[0] for desc in result.description]
gold_df = pd.DataFrame(gold_data, columns=columns)
print(gold_df)

# Close the connection
con.close()
```

#### Join two tables, from two databases on two columns

We are joining:
- `eth_emissions_silver` s and `eth_emissions_gold` g ON s.`total_net_emissions_eth` = g.`total` AND s.`datetime` = g.`datetime`

```python
# print out all columns from both eth_emissions_silver and eth_emissions_gold
print(silver_df.columns)

print(gold_df.columns)
```

```python
# Merge the dataframes based on the specified conditions
emissions_df = pd.merge(
    silver_df,
    gold_df,
    on=['datetime'],  # Join on the 'datetime' column
    how='inner'  # Inner join to keep only matching rows
)

# Filter rows where 'total_net_emission_eth' equals 'total'
emissions_df = emissions_df[emissions_df['total_net_emission_eth'] == emissions_df['total']]

# Drop the duplicated 'total' column (since they are equal)
emissions_df.drop(columns=['total'], inplace=True)

# Reset the index
emissions_df.reset_index(drop=True, inplace=True)

# Print the resulting dataframe
print(emissions_df)
print(emissions_df.columns)
```
